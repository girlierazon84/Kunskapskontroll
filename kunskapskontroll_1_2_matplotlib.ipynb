### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

# Svar:
`Object-Oriented approach` is recommended to use for more complicated plots and plots that are part of a bigger project.

`Pyplot / MATLAB approach` can be used for a quick interactive work.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

# Svar:




- **Figure**: The top-level container for plots.
- **Axes**: The individual plots or subplots within a figure.
- **Axis**: Number lines that define data limits and scaling.
- **Artist**: Elements in plots, like lines and text.

### When plotting in Matplotlib, what is the expected input data type?

# Svar:

The expected input data type for plotting in Matplotlib is usually `NumPy arrays (numpy.array)` or objects that can be passed to `numpy.asarray` such as **python lists** and **pandas data structures**. It's a common practice to transform these into NumPy array objects before creating plots for consistency and ease of use.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Object-Oriented Approach:
x = np.linspace(-4, 4, 100)
# print(x)
y = x ** 2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Object-Oriented Approach (y = x^2)')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
# Pyplot | MATLAB Approach.
plt.plot(x, y)
plt.title("Pyplot | MATLAB Approach (y = x^2)")
plt.xlabel('x')
plt.ylabel('y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x ** 2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Creating Subplots.
fig, axs = plt.subplots(1, 2, layout = 'constrained')
fig.suptitle('Different Plots', size = 30)

axs[0].scatter(random_data_x, random_data_y)
axs[0].set_title('Scatter Plot')

axs[1].bar(names, values)
axs[1].set_title('Bar Plot')
```
